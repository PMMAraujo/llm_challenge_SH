```python
import random
```

```python
def generate_first_enry(patient_id,name):
  initial_weight = random.randint(60, 100)
  initial_systolic_bp = random.randint(80, 130)
  initial_diastolic_bp = random.randint(60, 90)
  initial_date = [random.randint(1, 28), random.randint(1, 12), 23-number_of_observations]#f"{random.randint(1, 28)}-{random.randint(1, 12)}-{23-number_of_observations}"

  this_entry = {
        'patient_id': patient_id,
        'patient_name': name,
        'weight': initial_weight,
        'systolic_bp': initial_systolic_bp,
        'diastolic_bp': initial_diastolic_bp,
        'date': initial_date#f"{initial_date[0]}-{initial_date[1]}-{initial_date[2]}"
    }
  #return pd.DataFrame.from_dict(this_entry,orient='index')
  return this_entry

```

```python
def generate_subsequente_entreis(dict_with_first_entry):
  pid = dict_with_first_entry['patient_id']
  name = dict_with_first_entry['patient_name']
  current_weight = dict_with_first_entry['weight']
  current_systolic_bp = dict_with_first_entry['systolic_bp']
  current_diastolic_bp = dict_with_first_entry['diastolic_bp']
  current_date = dict_with_first_entry['date']

  lst_pid = [pid]
  lst_name = [name]
  lst_current_weight = [current_weight]
  lst_current_systolic_bp = [current_systolic_bp]
  lst_current_diastolic_bp = [current_diastolic_bp]
  lst_current_date = [f"{current_date[0]}-{current_date[1]}-{current_date[2]}"]

  number_of_observations = random.choice([3,5])

  for obs_n in range(number_of_observations -1):
    current_weight = random.randint(-10,10) + current_weight
    current_systolic_bp = random.randint(-5,5) + current_systolic_bp
    current_diastolic_bp = random.randint(-5,5) + current_diastolic_bp
    current_date = [current_date[0],current_date[1], current_date[2]+1]

    lst_pid.append(pid)
    lst_name.append(name)
    lst_current_weight.append(current_weight)
    lst_current_systolic_bp.append(current_systolic_bp)
    lst_current_diastolic_bp.append(current_diastolic_bp)
    lst_current_date.append(f"{current_date[0]}-{current_date[1]}-{current_date[2]}")

  df = pd.DataFrame()
  df['patient_id'] = lst_pid
  df['patient_name'] = lst_name
  df['weight'] = lst_current_weight
  df['systolic_bp'] = lst_current_systolic_bp
  df['diastolic_bp'] = lst_current_diastolic_bp
  df['date'] = lst_current_date

  return df
```

```python
first = generate_first_enry('PID00000', 'Eleanor Pires')
generate_subsequente_entreis(first)
```

```python
pairs_id_name = df[['patient_id','patient_name']].values
pairs_id_name[0]
```

```python
final_df = pd.DataFrame()

for id, name in pairs_id_name:
  first = generate_first_enry(id, name)
  df_patient = generate_subsequente_entreis(first)

  final_df = pd.concat([final_df, df_patient], ignore_index=True)
```

```python
final_df['date'] = pd.to_datetime(final_df['date'])
```

```python
final_df.set_index('patient_id', inplace=True)
```
